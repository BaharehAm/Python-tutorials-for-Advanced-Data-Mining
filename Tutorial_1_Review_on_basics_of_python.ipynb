## Lists

python-list.jpg

```python
## store multiple values with various data types in a list (lists allow duplicate values)
list1 = [20, 'Jessa', 35.75, [30,60,90]]
print(list1)

list2 = [20, 'Jessa', 'Jessa', 35.75, [30,60,90]]
print(list2)

## get items of a list (indexing starts from 0, slicing, negative index)
list2[0]
list2[2:4]
list2[-1]
list2[-4:-1]
list2[1:]
list2[4][1]


## change items (one item, a range of items)
list2[1] = 'python'
print(list2)

list2[2:4] = ["book", "fruit"]
print(list2)

list2[2:4] = [87]
print(list2)

list2[2:4] = [90, 100, 110]
print(list2)

## add items (.insert(), .append())
list2.insert(4,'python')
print(list2)

list3 = ['math', 'code']
list2.append(list3)
print(list2)

## extend lists (.extend(), "+" operation )
list4 = [0]
list2.extend(list4)
print(list2)

list5 = [9, 5]
list6 = [-3, 8]
list5 + list6

list5 = ['a', 'f']
list6 = ['d']
list5 + list6

## delete the whole list or specific items from the list
del list5
print(list5)

del list6[0]
print(list6)

## create list with range( , , )
# [0,1,2,3]
list7 = list(range(0,4))
list7 = list(range(4))
print(list7)

# [0, 2, 4, 6]
list8 = list(range(0,8,2))
print(list8)

```

## If statement

python-if-else.png
[Ref](https://www.programiz.com/python-programming/if-elif-else)

some example conditions:

1.   Equals: a == b
2.   Not Equals: a != b
3.   Less than: a < b
4.   Less than or equal to: a <= b
5.   Greater than: a > b
6.   Greater than or equal to: a >= b
7.   Combination of conditions with `AND`, `OR`


```python
## if
var10 = 57
if var10 < 40:
  print('var10 is less than 40')



## if ... else
var10 = 57
if var10 < 40:
  print('var10 is less than 40')
else:
  print('var10 is greater than 40')


## if ... elif ... elif ... else
var10 = 40
if var10 < 40:
  print('var10 is less than 40')
elif var10 == 40:
  print('var10 is equal to 40')
else:
  print('var10 is greater than 40')


## if ... if (nested if)
var10 = 45
if var10 > 40:
  if var10 > 50:
    print('var10 is greater than 50')
  else:
    print("var10 is not greater than 50")
else:
  print('var10 is less than 40')

```

##  Loops (for, while)

image.png
[Ref](https://www.scaler.com/topics/difference-between-for-and-while-loop-in-python/)

```python
## for loop for lists (iterate over items, iterate over indexes)
list11 = ['a', 2, 'b', 3]
for i in list11:
  print(i)


## for loop for a string
string1 = 'Python'
for ch in string1:
  print(ch)


## for ... for (nested loop)
list12 = [1, "o", "RR", 8.9]
list13 = [5, 0]

empty_list = []
for i in list12:
  for j in list13:
    empty_list.append([i,j])

print(empty_list)


## while -- condition
parameter = 5
while parameter > 2:
  parameter = parameter - 1 # 5-1=4 --> 4-1=3 --> 3-1=2
  print(parameter)


## never-ending loop
parameter = 5
while parameter > 2:
  print(parameter)
  break


## while ... if , for ... if
for i in list11:
  if i == 2:
    print(i)
  else:
    print(list11[-1])
# [3, 2, 3, 3]

```

## Functions

Python-Function-Syntax.png
[Ref](https://www.learnbyexample.org/python-functions/)

```python
## define a simple function with no arguments/parameters (def--return)
def add_two_values():
  2 + 3
  return 2+3


# def train_ANN(X, y, X_test, y_test num_units, epochs, batch_size):
#   model = ...
#   model.fit(X, y, epochs = epochs , batch_size=batch_size)
#   accuracy = model.evaluate(X_test, y_test)
#   return accuracy



## call a function
add_two_values()

## define a function with arguments/parameters
def add_two_values(a, b):
  c = a + b
  return c

add_two_values(5, 10)

add_two_values(10, 5)


## set a default value for the parameter
def add_two_values(a, b=2):
  c = a + b
  return c
add_two_values(7, 6)
add_two_values(7)


def add_two_values(a=4, b=2):
  c = a + b
  return c
add_two_values()


## order of parameters (right order, any order by using keys/values)
add_two_values(5, 10)
add_two_values(b=10, a=5)

```
